```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Define the URL
url = 'https://www.imdb.com/chart/top'

# Fetch the page content
# Modify the headers to include a User-Agent string
request_headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Initiate HTTP request
response = requests.get(url, headers=request_headers)

soup = BeautifulSoup(response.text, 'html.parser')

```

```python
movie_list = soup.find("ul", {"class": "ipc-metadata-list ipc-metadata-list--dividers-between sc-a1e81754-0 eBRbsI compact-list-view ipc-metadata-list--base"})
```

```python
movies = movie_list.find_all("li", {"class": "ipc-metadata-list-summary-item sc-10233bc-0 iherUv cli-parent"})
```

```python
data = []
for movie in movies:
  data_div = movie.find("div", {"class": "ipc-metadata-list-summary-item__c"})
  title = data_div.find("h3", {"class": "ipc-title__text"}).text[2:].strip()
  meta_data = data_div.find_all("span", {"class": "sc-b0691f29-8 ilsLEX cli-title-metadata-item"})
  year = meta_data[0].text.strip()
  duration = meta_data[1].text.strip()
  data.append([title, year, duration])


```

```python
df = pd.DataFrame(data)
```

```python
df.columns = ["title", "year", "duration"]
```
